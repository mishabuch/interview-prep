```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
classifier(["I've been waiting for a HuggingFace course my whole life.","I love taking AI courses", "I hate long workouts though"])

classifier_zeroshot = pipeline("zero-shot-classification")
classifier_zeroshot(
    ["This is a course about the Transformers library","BJP is a perfectly nationalistic party"],
    candidate_labels=["education", "politics", "business"],
)

unmasker = pipeline("fill-mask")
unmasker("This course will teach you all about <mask> models.", top_k=2)

ner = pipeline("ner")
ner("My name is Sylvain and I work at Hugging Face in Brooklyn.")

transcriber = pipeline(
    task="automatic-speech-recognition", model="openai/whisper-large-v3"
)
result = transcriber(
    "https://huggingface.co/datasets/Narsil/asr_dummy/resolve/main/mlk.flac"
)
print(result)
```
